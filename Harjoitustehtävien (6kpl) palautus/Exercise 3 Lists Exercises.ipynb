```python
#Harjoitus 2, kohta 1 & 2
import random
thrownDiceNumbers=[]

for i in range(5):
    thrownDiceNumbers.append(random.randint(1, 6))
```

```python
#Harjoitus 2, kohta 3
print(thrownDiceNumbers)
```

```python
#Harjoitus 2, kohta 4
total = sum(thrownDiceNumbers)
print("The value of numbers:", total)
```

```python
#Harjoitus 2, kohta 5
highest = max(thrownDiceNumbers)
print("Highest value:", highest)
```
